# <span style="color:orange">Apriori</span>

### <span style="color:orange">Importing the Libraries</span>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### <span style="color:orange">Importing the Dataset</span>

```python
dataset = pd.read_csv('Dataset\\Market_Basket_Optimisation.csv', header = None)

dataset
```

### <span style="color:orange">Data Preprocessing</span>

```python
transactions = []
for row in range(0, 7501):
    transactions.append([str(dataset.values[row, col] ) for col in range(0, 20) ] )
```

### <span style="color:orange">Training the Apriori Model on the Dataset</span>

```python
from apyori import apriori
```

```python
rules = apriori(transactions=transactions, min_support=0.003, min_confidence=0.2, min_lift=3,
                min_length=2, max_length=2)
```

### <span style="color:orange">Visualizing the Results</span>

#### <span style="color:orange">Displaying the first results coming directly from the output of the Apriori function</span>

```python
print(type(rules) )
```

```python
print(rules)
```

```python
results = list(rules)

results
```

#### <span style="color:orange">Putting the results well organized into a Pandas Dataframe</span>

```python
def inspect(results):
    lhs = [tuple(result[2][0][0]) [0] for result in results]
    rhs = [tuple(result[2][0][1]) [0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts) )
results_in_dataframe = pd.DataFrame(inspect(results), columns=['Left Hand Side', 'Right Hand Side', 'Support',
                                                      'Confidence', 'Lift'])
```

##### <span style="color:orange">Displaying the results non sorted</span>

```python
results_in_dataframe
```

##### <span style="color:orange">Displaying the results sorted by descending lifts</span>

```python
results_in_dataframe.nlargest(n=10, columns='Lift')
```
