### <span style="color:orange">Importing the Libraries</span>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
import pandas as pd
```

### <span style="color:orange">Importing the dataset</span>

```python
dataset = pd.read_csv('Datasets//Position_Salaries.csv')

input_feature = dataset.iloc[ : , 1:-1].values
dependent_variable = dataset.iloc[ : , -1].values
```

### <span style="color:orange">Training the Decision Tree Regression Model</span>

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
decision_tree_regressor = DecisionTreeRegressor(random_state=0)
```

```python
decision_tree_regressor.fit(X=input_feature, y=dependent_variable)
```

### <span style="color:orange">Predicting a New Result</span>

```python
print(decision_tree_regressor.predict(X=[[6.5] ] ) )  # // remember the prediction is in 2 sq. brackets
```

### <span style="color:orange">Visualizing the results</span>

**Remember that this model wasn't a good fit for this dataset**

```python
## real points
plt.scatter(input_feature, dependent_variable, color="orange")
## decision tree prediction (plotting the line of the DT predictions)
plt.plot(input_feature, decision_tree_regressor.predict(input_feature), color="blue")

plt.title("Decision Tree predictions with the real values")
plt.xlabel("Position Level.")
plt.ylabel("Salary.")

plt.show()
```
