# <span style="color:orange">Random Forest</span>

### <span style="color:orange">Importing the Libraries</span>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### <span style="color:orange">Importing the Dataset</span>

```python
dataset = pd.read_csv('Dataset//Position_Salaries.csv')

input_feature = dataset.iloc[ : , 1:-1].values
dependent_variable = dataset.iloc[ : , -1].values
```

```python
dataset
```

```python
print(input_feature)
```

```python
print(dependent_variable)
```

### <span style="color:orange">Training the Random Forest Regression Model</span>

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random_forest_regressor = RandomForestRegressor(n_estimators=10, random_state=0)

random_forest_regressor.fit(X=input_feature, y=dependent_variable)
```

### <span style="color:orange">Predicting a New Result</span>

```python
print(random_forest_regressor.predict([[6.5]] ) ) # // 2D array -> 2 sq brackets
```

### <span style="color:orange">Visualizing the Results</span>

```python
## real points
plt.scatter(input_feature, dependent_variable, color="orange")
## decision tree prediction (plotting the line of the DT predictions)
plt.plot(input_feature, random_forest_regressor.predict(input_feature), color="blue")

plt.title("Random Forest predictions with the real values")
plt.xlabel("Position Level.")
plt.ylabel("Salary.")

plt.show()
```
