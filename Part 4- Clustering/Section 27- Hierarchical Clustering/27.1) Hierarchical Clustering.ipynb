# <span style="color:orange">Hierarchical Clustering</span>

### <span style="color:orange">Importing the Libraries</span>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.cluster.hierarchy import dendrogram
```

### <span style="color:orange">Importing the Dataset</span>

```python
dataset = pd.read_csv('Dataset\\Mall_Customers.csv')

dataset
```

```python
input_features = dataset.iloc[ : , [3, 4] ].values
```

### <span style="color:orange">Using the Dendrogram to find the Optimal Number of Clusters</span>

```python
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(input_features, method='ward') )
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

### <span style="color:orange">Training the Hierarchical Clustering model on the dataset</span>

```python
from sklearn.cluster import AgglomerativeClustering

hiearchical_cluster = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')

dv_hiearchical = hiearchical_cluster.fit_predict(input_features)
```

```python
print(dv_hiearchical)
```

### <span style="color:orange">Visualizing the Clusters</span>

```python
plt.figure(figsize=(10, 6) )
plt.scatter(input_features[dv_hiearchical == 0, 0], input_features[dv_hiearchical == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(input_features[dv_hiearchical == 1, 0], input_features[dv_hiearchical == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(input_features[dv_hiearchical == 2, 0], input_features[dv_hiearchical == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(input_features[dv_hiearchical == 3, 0], input_features[dv_hiearchical == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(input_features[dv_hiearchical == 4, 0], input_features[dv_hiearchical == 4, 1], s=100, c='magenta', label='Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
