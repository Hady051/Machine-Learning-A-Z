### <span style="color:orange">Importing the Libraries</span>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import logistic
```

### <span style="color:orange">Importing the Dataset</span>

```python
dataset = pd.read_csv('Dataset\\Social_Network_Ads.csv')
```

```python
dataset
```

```python
print(dataset)
```

```python
input_features = dataset.iloc[ : , :-1].values
dv = dataset.iloc[ : , -1].values
```

```python
print(input_features)
```

```python
print(dv)
```

### <span style="color:orange">Splitting the Data</span>

```python
from sklearn.model_selection import train_test_split

(input_features_train, input_features_test,
dv_train, dv_test) = train_test_split(input_features, dv, test_size=0.2, random_state=0)
```

```python
print(input_features_train)
```

### <span style="color:orange">Feature Scaling</span>

##### <span style="color:orange">Standardization</span>

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
```

```python
## X-train
input_features_train = standard_scaler.fit_transform(input_features_train)

## X-test transform only
input_features_test = standard_scaler.transform(input_features_test)
```

```python
print(input_features_train)
```

### <span style="color:orange">Training the Logistic Regression Model on the Training set</span>

```python
from sklearn.linear_model import LogisticRegression

## Building the Model
logistic_R_classifier = LogisticRegression(random_state=0)
## Training the Model
logistic_R_classifier.fit(X=input_features_train, y=dv_train)
```

### <span style="color:orange">Predicting a New Result</span>

```python
## Predicting a new customer who is 30 years old with an estimated salary of $87,000
new_prediction = logistic_R_classifier.predict(standard_scaler.transform([[30, 87000]] ) )

buy_prediction = ""
if new_prediction == 1:
    buy_prediction = 'buy'
elif new_prediction == 0:
    buy_prediction = 'not buy'

print("This customer is predicted to: ", buy_prediction)
```

```python
new_prediction_prob = logistic_R_classifier.predict_proba(standard_scaler.transform([[30, 87000]] ) )

print(f"The probability of the new customer buying is: {new_prediction_prob[0][1]:.2%}")
```

### <span style="color:orange">Predicting the Test Set Results</span>

```python
dv_predicted = logistic_R_classifier.predict(input_features_test)
```

```python
np.set_printoptions(precision=2)

print(np.concatenate((dv_predicted.reshape(len(dv_predicted), 1),
                      dv_test.reshape(len(dv_test), 1) ), axis=1) )
# // [dv_predicted dv_test]
```

### <span style="color:orange">**Making the Confusion Matrix and getting the Evaluation scores</span>

```python
from sklearn.metrics import confusion_matrix

confusion_matrix_score = confusion_matrix(y_true=dv_test, y_pred=dv_predicted)
print(confusion_matrix_score)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score = accuracy_score(y_true=dv_test, y_pred=dv_predicted)
print(accuracy_score)
```

```python
from sklearn.metrics import precision_score

precision_score = precision_score(y_true=dv_test, y_pred=dv_predicted)
print(precision_score)
```

```python
from sklearn.metrics import recall_score

recall_score = recall_score(y_true=dv_test, y_pred=dv_predicted)
print(recall_score)
```

```python
from sklearn.metrics import f1_score

f1_score = f1_score(y_true=dv_test, y_pred=dv_predicted)
print(f1_score)
```

### <span style="color:orange">Visualizing the Training Set Results</span>

```python
from matplotlib.colors import ListedColormap

input_features_set, dv_set = standard_scaler.inverse_transform(input_features_train), dv_train
X1, X2 = np.meshgrid(np.arange(start=input_features_set[:, 0].min() - 10,
                               stop=input_features_set[:, 0].max() + 10,
                               step=0.25),
                     np.arange(start=input_features_set[:, 1].min() - 1000,
                               stop=input_features_set[:, 1].max() + 1000,
                               step=0.25) )
plt.contourf(X1, X2,
             logistic_R_classifier.predict(standard_scaler.transform(np.array(
                 [X1.ravel(), X2.ravel()] ).T) ).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(('red', 'green') ) )
plt.xlim(X1.min(), X1.max() )
plt.ylim(X2.min(), X2.max() )

for row, col in enumerate(np.unique(dv_set)):
    plt.scatter(input_features_set[dv_set == col, 0],
                input_features_set[dv_set == col, 1],
                color = ListedColormap(('red', 'green') )(row),
                label = f'class[{col}]')

plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

### <span style="color:orange">Visualizing the Test Set Results</span>

```python
from matplotlib.colors import ListedColormap

input_features_set, dv_set = standard_scaler.inverse_transform(input_features_test), dv_test
X1, X2 = np.meshgrid(np.arange(start=input_features_set[:, 0].min() - 10,
                               stop=input_features_set[:, 0].max() + 10,
                               step=0.25),
                     np.arange(start=input_features_set[:, 1].min() - 1000,
                               stop=input_features_set[:, 1].max() + 1000,
                               step=0.25) )

plt.contourf(X1, X2,
             logistic_R_classifier.predict(standard_scaler.transform(np.array(
                 [X1.ravel(), X2.ravel()] ).T) ).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(('red', 'green') ) )
plt.xlim(X1.min(), X1.max() )
plt.ylim(X2.min(), X2.max() )

for row, col in enumerate(np.unique(dv_set)):
    plt.scatter(input_features_set[dv_set == col, 0],
                input_features_set[dv_set == col, 1],
                color = ListedColormap(('red', 'green') )(row),
                label = f'class[{col}]')

plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
