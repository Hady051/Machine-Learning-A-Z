### <span style="color:orange">Importing the Libraries</span>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### <span style="color:orange">Importing the Dataset</span>

```python
dataset = pd.read_csv('Dataset\\Social_Network_Ads.csv')
```

```python
dataset
```

```python
print(dataset)
```

```python
input_features = dataset.iloc[ : , :-1].values
dv = dataset.iloc[ : , -1].values
```

```python
print(input_features)
```

```python
print(dv)
```

### <span style="color:orange">Splitting the Data</span>

```python
from sklearn.model_selection import train_test_split

(input_features_train, input_features_test,
dv_train, dv_test) = train_test_split(input_features, dv, test_size=0.2, random_state=0)
```

```python
print(input_features_train)
```

### <span style="color:orange">Feature Scaling</span>

##### <span style="color:orange">Standardization</span>

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
```

```python
## X-train
input_features_train = standard_scaler.fit_transform(input_features_train)

## X-test transform only
input_features_test = standard_scaler.transform(input_features_test)
```

```python
print(input_features_train)
```

### <span style="color:orange">Training the Random Forest Model on the Training set</span>

```python
from sklearn.ensemble import RandomForestClassifier
#
# Building the Model
random_F_classifier = RandomForestClassifier(n_estimators=10, random_state=0)
# Training the Model
random_F_classifier.fit(X=input_features_train, y=dv_train)
```

### <span style="color:orange">* The exhaustive grid search to optimize the hyperparameters</span>

```python
## Preparing the parameter grid:

n_estimators = [int(x) for x in np.linspace(start = 10, stop = 200, num = 10)]
criterion = ['gini', 'entropy']
max_features = ['auto', 'sqrt', 'log2']
max_depth = [int(x) for x in np.linspace(10, 100, num = 10)]
min_samples_split = [2, 4, 8, 16]
min_samples_leaf = [2, 4, 8, 16]
bootstrap = [True, False]

param_grid = {'n_estimators': n_estimators,
                'criterion': criterion,
                'max_features': max_features,
                'max_depth': max_depth,
                'min_samples_split': min_samples_split,
                'min_samples_leaf': min_samples_leaf,
                'bootstrap': bootstrap}
```

****I didn't continue it because it takes a long time, so I took the parameters of the poster**

```python
## Use best params from the grid search
from sklearn.ensemble import RandomForestClassifier

# Building the model with the best parameters
random_F_classifier_grid_S = RandomForestClassifier(n_estimators=10,
                                        criterion='gini',
                                        bootstrap=True,
                                        max_depth=10,
                                        max_features='sqrt',
                                        min_samples_leaf=8,
                                        min_samples_split=2,
                                        random_state = 0)
# Training the model
random_F_classifier_grid_S.fit(input_features_train, dv_train)
```

### <span style="color:orange">Predicting a New Result</span>

```python
## Predicting a new customer who is 30 years old with an estimated salary of $87,000
new_prediction = random_F_classifier_grid_S.predict(standard_scaler.transform([[30, 87000]] ) )

buy_prediction = ""
if new_prediction == 1:
    buy_prediction = 'buy'
elif new_prediction == 0:
    buy_prediction = 'not buy'

print("This customer is predicted to: ", buy_prediction)
```

### <span style="color:orange">Predicting the Test Set Results</span>

```python
dv_predicted = random_F_classifier_grid_S.predict(input_features_test)
```

```python
np.set_printoptions(precision=2)

print(np.concatenate((dv_predicted.reshape(len(dv_predicted), 1),
                      dv_test.reshape(len(dv_test), 1) ), axis=1) )
# // [dv_predicted dv_test]
```

### <span style="color:orange">**Making the Confusion Matrix and getting the Evaluation scores</span>

```python
from sklearn.metrics import confusion_matrix

confusion_matrix_score = confusion_matrix(y_true=dv_test, y_pred=dv_predicted)
print(confusion_matrix_score)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score = accuracy_score(y_true=dv_test, y_pred=dv_predicted)
print(accuracy_score)
```

```python
from sklearn.metrics import precision_score

precision_score = precision_score(y_true=dv_test, y_pred=dv_predicted)
print(precision_score)
```

```python
from sklearn.metrics import recall_score

recall_score = recall_score(y_true=dv_test, y_pred=dv_predicted)
print(recall_score)
```

```python
from sklearn.metrics import f1_score

f1_score = f1_score(y_true=dv_test, y_pred=dv_predicted)
print(f1_score)
```
